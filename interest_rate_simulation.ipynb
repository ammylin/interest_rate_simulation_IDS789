```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
initial_rate = 0.05             # Starting interest rate (5%)
mean = 0                        # Mean of underlying normal distribution
sigma = 0.00625                 # Daily standard deviation
days = 90                       # 3 months = 90 days
num_paths = 10                  # Simulate 10 scenarios

# Set seed for reproducibility
np.random.seed(42)
all_paths = [] 
for i in range(num_paths):
    rate_changes = np.random.lognormal(mean, sigma, days)
    rates = [initial_rate]

    for day in range(days): 
        new_rate = rates[-1] * rate_changes[day]
        rates.append(new_rate)
    all_paths.append(rates)

plt.figure(figsize=(12, 6))
for i, path in enumerate(all_paths):
    plt.plot(path, label=f'Path {i+1}')
    
plt.title("Simulated 3-Month Interest Rate Paths Using np.random.lognormal()")
plt.xlabel("Day")
plt.ylabel("Interest Rate")
plt.grid(True)
plt.tight_layout()
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
